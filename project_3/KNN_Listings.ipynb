I am reassigning the dataframe into a new variable since I redimensioned the size sort to speak.

```python
# new_data = df.dropna(axis = 0, how ='any') 
new_data = df
new_data
```

Again this is for me but I want to display the columsn that is in the new dataframe and rearrange

```python
[col for col in new_data.columns]
```

This is where I rearrange the columns 

```python
col_rearrange = ['latitude',
 'longitude','accommodates',
 'bathrooms','bedrooms',
 'beds', 'minimum_nights',
 'maximum_nights','price']

new_data = new_data.reindex(columns=col_rearrange)
```

I want to display the dataframe with the rearranged columns to see if I have done it correctly.

```python
new_data
```

This is where I want to convert the column price to float and remove any extraneous symbols that might 
cause a compiling error when run.

```python
new_data['price'] = new_data['price'].replace({'\$': '', ',': ''}, regex=True).astype(float)
```

Here I want to drop any null or zero values in all of the columns

```python
zero_not_accepted = ['accommodates','bathrooms','bedrooms','beds','minimum_nights','maximum_nights','price']

for column in zero_not_accepted:
    new_data[column] = new_data[column].replace(0, np.NaN)
    mean = int(new_data[column].mean(skipna=True))
    new_data[column] = new_data[column].replace(np.NaN, mean)
```

Displaying the dataframe again so I see I dropped the $ symbols and any other symbols that might interfere with 
using KNN algorithm

```python
new_data
```

Here I am setting up my X and y variables from the dataframe.  My X values will consist of values from columns 0 to 5 and my y values will contains columns from 6 which is price.

```python
X = new_data.iloc[:, 0:8]
y = new_data.iloc[:, 8]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

From what I understand, this is putting the X values into the X_train variable to train it and the X values to transform into te X_test.

```python
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

Here I will define the model I am using which is the K-NN model.

```python
classifier = KNeighborsClassifier(n_neighbors=11, p=2, metric='euclidean')
```

```python
# Fit model
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(f1_score(y_test, y_pred , average='micro'))
```

```python
print(accuracy_score(y_test, y_pred))
```
